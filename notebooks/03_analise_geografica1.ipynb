# Notebook 03 — Análise Geográfica (Mapas por Bairro)

```
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd
```

```
mapa = gpd.read_file('analise-perfil-eleitor-vila-velha/dados/geodata/bairros_vila_velha.shp')
mapa.head()
```

```
base = pd.read_csv('analise-perfil-eleitor-vila-velha/dados/base_vila_velha.csv')
base.head()
```

```
votos_bairro = base.groupby('BAIRRO')['QTD_ELEITORES'].sum().reset_index()
votos_bairro
```

```
mapa = mapa.merge(votos_bairro, left_on='nome_bairro', right_on='BAIRRO', how='left')
mapa.head()
```

```
fig, ax = plt.subplots(figsize=(10, 10))
mapa.plot(column='QTD_ELEITORES', legend=True, ax=ax)
plt.title('Distribuição de Eleitores por Bairro — Vila Velha')
plt.show()
```
