```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

df = pd.read_csv('dados/base_vila_velha.csv')
os.makedirs('docs', exist_ok=True)
```

## Distribui√ß√£o por Faixa Et√°ria

```python
plt.figure(figsize=(10,5))
sns.histplot(df_vv['faixa_etaria'], bins=20)
plt.title('Distribui√ß√£o por Faixa Et√°ria')
plt.savefig('docs/faixa_etaria.png')
plt.show()
```

## Distribui√ß√£o por Sexo

```python
df_vv['genero'].value_counts().plot(kind='bar')
plt.title('Distribui√ß√£o por Sexo')
plt.savefig('docs/sexo.png')
plt.show()
```

## Distribui√ß√£o por Grau de Instru√ß√£o

```python
df_vv['escolaridade'].value_counts().plot(kind='bar', figsize=(10,5))
plt.title('Grau de Instru√ß√£o')
plt.savefig('docs/grau_instrucao.png')
plt.show()
```

## Eleitores por Bairro

```python
df_vv.groupby('bairro').size().sort_values(ascending=False).head(20).plot(kind='bar', figsize=(12,5))
plt.title('Top 20 Bairros por N√∫mero de Eleitores')
plt.savefig('docs/bairros_top20.png')
plt.show()
```

## Eleitores por Regi√£o

```python
df_vv['regiao'].value_counts().sort_index().plot(kind='bar')
plt.title('Eleitores por Regi√£o')
plt.savefig('docs/regioes.png')
plt.show()
```

## Eleitores por Zona Eleitoral

```python
df_vv.groupby('zona').size().plot(kind='bar')
plt.title('Eleitores por Zona Eleitoral')
plt.savefig('docs/zonas.png')
plt.show()
```
