Fun√ß√£o auxiliar para rotular Barras

````python
def add_labels(ax):
    """Adiciona r√≥tulos de valores no topo das barras."""
    for p in ax.patches:
        ax.annotate(
            f"{int(p.get_height()):,}".replace(",", "."),
            (p.get_x() + p.get_width()/2, p.get_height()*1.005),
            ha='center',
            fontsize=9
        )
````

Gerar Insitghs por BAIRRO

````python
def gerar_insights_bairro(df, bairro):
    # Filtrar o bairro solicitado
    df_b = df_vv[df["bairro"].str.lower() == bairro.lower()]

    if df_b.empty:
        return f"‚ùå Bairro '{bairro}' n√£o encontrado na base."

    # Total de eleitores
    total_eleitores = df_b.shape[0]

    # Total de eleitores
    local_votacao_nome = df_b.shape[0]

    # Faixa et√°ria predominante
    faixa_pred = df_b['faixa_etaria'].mode()[0]

    # G√™nero predominante
    genero_pred = df_b['genero'].mode()[0]

    # Escolaridade predominante
    escolar_pred = df_b['escolaridade'].mode()[0]

    # Zona eleitoral predominante
    zona_pred = df_b['zona'].mode()[0]

    # Regi√£o (num_regiao)
    regiao = df_b['regiao'].mode()[0]

    # Top faixas et√°rias (contagem)
    top_faixas = df_b['faixa_etaria'].value_counts().head(5)

    # Retornar os insights em dicion√°rio
    insights = {
        "Bairro": bairro.title(),
        "Regi√£o": regiao,
        "Total de eleitores": total_eleitores,
        "Local de Vota√ß√£o": local_votacao_nome,
        "Faixa et√°ria predominante": faixa_pred,
        "G√™nero predominante": genero_pred,
        "Escolaridade predominante": escolar_pred,
        "Zona eleitoral predominante": zona_pred,
        "Top faixas et√°rias": top_faixas.to_dict()
    }

    return insights
````

````python
def texto_bairro(insights):
    texto = (
        f"üìç **Bairro {insights['Bairro']} ‚Äî Regi√£o {insights['Regi√£o']}**\n\n"
        f"O bairro possui **{insights['Total de eleitores']} eleitores** cadastrados. "
        f"O local de vota√ß√£o √© **{insights['Local de Vota√ß√£o']}**. \n\n"
        f"A faixa et√°ria predominante √© **{insights['Faixa et√°ria predominante']}**, "
        f"indicando o principal perfil et√°rio da regi√£o.\n\n"
        f"O g√™nero mais frequente √© **{insights['G√™nero predominante']}**, "
        f"e a escolaridade mais comum entre os moradores √© **{insights['Escolaridade predominante']}**.\n\n"
        f"Al√©m disso, a zona eleitoral mais recorrente para os eleitores deste bairro √© a **{insights['Zona eleitoral predominante']}**.\n"
    )
    return texto
````

````python
def gerar_insights_bairro(df_vv, bairro):
    bairro_upper = bairro.upper().strip()

    # Filtrar bairro
    df_bairro = df_vv[df_vv["bairro"].str.upper() == bairro_upper]

    if df_bairro.empty:
        return None  # Retorna None se o bairro n√£o for encontrado

    # Total de eleitores
    total_eleitores = df_bairro["quant_eleitores"].sum()

    # Locais de vota√ß√£o presentes no bairro
    locais = df_bairro["local_votacao_nome"].unique()
    total_locais = len(locais)

    # Eleitores por local de vota√ß√£o
    eleitores_por_local = (
        df_bairro.groupby("local_votacao_nome")["quant_eleitores"].sum().sort_values(ascending=False)
    )

    # Faixa et√°ria
    faixa_etaria_dist = (
        df_bairro.groupby("faixa_etaria")["quant_eleitores"].sum().sort_values(ascending=False)
    )

    # Escolaridade
    escolaridade_dist = (
        df_bairro.groupby("escolaridade")["quant_eleitores"].sum().sort_values(ascending=False)
    )

    # Sexo
    sexo_dist = (
        df_bairro.groupby("genero")["quant_eleitores"].sum()
    )

    # Faixa et√°ria predominante
    faixa_pred = faixa_etaria_dist.index[0] if not faixa_etaria_dist.empty else "N√£o Informado"

    # G√™nero predominante
    genero_pred = sexo_dist.idxmax() if not sexo_dist.empty else "N√£o Informado"

    # Escolaridade predominante
    escolar_pred = escolaridade_dist.index[0] if not escolaridade_dist.empty else "N√£o Informado"

    # Zona eleitoral predominante
    zona_pred = df_bairro['zona'].mode()[0] if not df_bairro['zona'].empty else "N√£o Informado"

    # Regi√£o
    regiao = df_bairro['regiao'].mode()[0] if not df_bairro['regiao'].empty else "N√£o Informado"

    return {
        "bairro_nome": bairro.title(),
        "total_eleitores": total_eleitores,
        "total_locais_votacao": total_locais,
        "locais_votacao_nomes": locais.tolist(),
        "eleitores_por_local": eleitores_por_local.to_dict(),
        "faixa_etaria_dist": faixa_etaria_dist.to_dict(),
        "escolaridade_dist": escolaridade_dist.to_dict(),
        "sexo_dist": sexo_dist.to_dict(),
        "faixa_etaria_predominante": faixa_pred,
        "genero_predominante": genero_pred,
        "escolaridade_predominante": escolar_pred,
        "zona_eleitoral_predominante": zona_pred,
        "regiao": regiao
    }
````
